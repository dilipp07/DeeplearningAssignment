# To start with deep learning, the very basic project that you can build is to predict the next digit in a sequence.

```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np

# Create the sequence of odd numbers from 1 to 99
sequence = np.arange(1, 100, 2)
```

```python
input_data = sequence[:-1]
output_data = sequence[1:]
```

```python
input_data
```

```python
# Define the neural network model
model = keras.Sequential([
    keras.layers.Dense(16, input_shape=(1,), activation='relu'),
    keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 16, kernel_initializer = 'he_uniform',activation='relu',input_dim = 1))

# Adding the second hidden layer
classifier.add(Dense(units = 16, kernel_initializer = 'he_uniform',activation='relu'))

# Adding the output layer
classifier.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='linear'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'mse')
```

```python
# Fitting the ANN to the Training set
model_history=classifier.fit(input_data, output_data,validation_split=0.10, batch_size = 1, epochs = 200)
```

```python
# Make predictions
next_digit = classifier.predict(np.array([107]))
print(int(next_digit[0,0]))  # The predicted next digit
```
